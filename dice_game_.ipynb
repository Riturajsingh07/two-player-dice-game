```python
import random
print("------> Welcome To The Dice game <-------")
p1 = 0
p2 = 0
while p1<=20 and p2<=20:
  print(" Player 1 turn :")
  a = input(" press  enter for dice roll: ")
  dice = random.randint(1,6)
  p1 = p1+dice
  print("player 1 on stair :",p1)
  if p1>=20:
    break
  print(" player 2 turn :")
  a = input("press enter for dice roll :")
  dice = random.randint(1,6)
  p2 = p2+dice
  print("player 2 on stair :",p2)
  if p2>=20 :
    break
if p1>p2 :
      print(" player 1  win ")
elif p1 == p2 :
      print("Match tie")
else:
      print(" player 2 win")




```
